```python
import torch
from torch import nn
```

```python
N = 10
A = nn.Embedding(N, 3)
```

```python
print(A.weight)
```

```python
full_range = torch.arange(N)
targ = 2
context_indices = full_range
k_ind = torch.argsort(torch.sum(torch.pow(A(context_indices) - A(torch.tensor(targ)), 2), dim = -1))
```

```python
import string
vocabulary = {}
for index, letter in enumerate(string.ascii_lowercase[:N]):
    vocabulary[letter] = index
```

```python
k_ind
```

```python
for i in k_ind:
    word = list(vocabulary.keys())[i]
    print(word, vocabulary[word])
```

```python
window_size = 2
idx = 0
N_max = 4
```

```python
for idx in range(N_max):
    start = idx - window_size
    end = idx + window_size + 1
    full_range = torch.arange(start, end)
    print(full_range)
    negative_left_elements = (full_range < 0)
    negative_right_elements = (full_range >= N_max)
    full_range[negative_left_elements] = -full_range[negative_left_elements] + 2*idx
    full_range[negative_right_elements] -= (end - start - 1)
    print(full_range)
```
